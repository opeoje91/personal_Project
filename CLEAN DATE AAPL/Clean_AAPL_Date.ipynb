```python
# Initial imports.
import pandas as pd
from path import Path
```

```python
# Loading data after database join of seasons/quarters file and Apple's stock file
# https://towardsdatascience.com/4-tricks-you-should-know-to-parse-date-columns-with-pandas-read-csv-27355bb2ad0e#:~:text=By%20default%2C%20date%20columns%20are%20parsed%20using%20the,a%20different%20date%20format%2C%20for%20example%2C%20YYYY-DD-MM%20HH%3AMM%3ASS%3A
file_path =Path("AAPL.csv")
aapl_df = pd.read_csv(file_path, parse_dates=['Date'])
aapl_df.head(10)
```

```python
#apply a filter on your data, for the first 10 columns :
aapl_no_date = aapl_df.iloc[:,[1,2,3,4,5,6]]
```

```python
 aapl_no_date.to_csv('AAPL_NO_DATE.csv')
```

```python
file_path =Path("AAPL.csv")
aapl2_df = pd.read_csv(file_path, parse_dates=['Date'])
aapl2_df.head(10)
```

```python
#apply a filter on your data, for the first  columns :
aapl2_df = aapl2_df.iloc[:,[0]]
```

```python
aapl2_df.to_csv('AAPL_DATE.csv')
```

```python
formatted_AAPL2_df = aapl2_df["Date"].dt.strftime("%m/%d/%y")

print(formatted_AAPL2_df)
formatted_AAPL2_df.to_csv('CLEAN_AAPL_DATE.csv')
```

```python
file_path =Path("CLEAN_AAPL_DATE.csv")
aapl_clean_date_df = pd.read_csv(file_path, parse_dates=['Date'])
aapl_clean_date_df.set_index('Unnamed: 0',inplace=True)
aapl_clean_date_df.index.name=None
print(aapl_clean_date_df)
```

```python
file_path =Path("AAPL_NO_DATE.csv")
aapl_no_date_df = pd.read_csv(file_path)
aapl_no_date_df.set_index('Unnamed: 0',inplace=True)
aapl_no_date_df.index.name=None
print(aapl_no_date_df)
```

```python
# merge both
merged_CLEAN_DATE_AAPL = aapl_clean_date_df.join(aapl_no_date_df, how='outer')
merged_CLEAN_DATE_AAPL.head()
```

```python
print(merged_CLEAN_DATE_AAPL)
```

```python
# save to csv
merged_CLEAN_DATE_AAPL.to_csv('merged_CLEAN_DATE_AAPL.csv')
```
