```python
# Initial imports.
import pandas as pd
from path import Path
```

```python
# Loading data after database join of seasons/quarters file and Apple's stock file
# https://towardsdatascience.com/4-tricks-you-should-know-to-parse-date-columns-with-pandas-read-csv-27355bb2ad0e#:~:text=By%20default%2C%20date%20columns%20are%20parsed%20using%20the,a%20different%20date%20format%2C%20for%20example%2C%20YYYY-DD-MM%20HH%3AMM%3ASS%3A
file_path =Path("AAPL_Mock_ML_CLOSE_DATE.csv")
date_df = pd.read_csv(file_path, parse_dates=['Date'])
date_df.set_index('Unnamed: 0',inplace=True)
date_df.index.name=None
date_df.head()
```

```python
file_path =Path("AAPL_Mock_ML_CLOSE_NO_DATE.csv")
no_date_df = pd.read_csv(file_path)
no_date_df.set_index('Unnamed: 0',inplace=True)
no_date_df.index.name=None
no_date_df.head()
```

```python
# merge both
merged_AAPL_MOCK_ML_CLOSE = date_df.join(no_date_df, how='outer')
merged_AAPL_MOCK_ML_CLOSE.head()
```

```python
# save to csv
merged_AAPL_MOCK_ML_CLOSE.to_csv('merged_AAPL_MOCK_ML_CLOSE.csv')
```
